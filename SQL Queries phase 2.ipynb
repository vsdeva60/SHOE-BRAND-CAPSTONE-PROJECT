# Q.N2

```python
%%sql
SELECT category, AVG(comfort_rating) AS average_comfort_rating
FROM (
    SELECT category, comfort_rating
    FROM table1

    UNION ALL

    SELECT category, comfort_rating
    FROM table3
) AS combined_data
GROUP BY category;
```

# Q.N3

```python
%%sql
WITH CombinedData AS (
    SELECT product_id, performance_rating
    FROM table1

    UNION ALL

    SELECT product_id, performance_rating
    FROM table3
)

SELECT product_id, performance_rating
FROM CombinedData
WHERE performance_rating > (SELECT AVG(performance_rating) FROM CombinedData);
```

# Q.N4

```python
%%sql
WITH CombinedData AS (
    SELECT product_id, comfort_rating
    FROM table1

    UNION ALL

    SELECT product_id, comfort_rating
    FROM table3
)

SELECT product_id, comfort_rating
FROM CombinedData
WHERE comfort_rating > (SELECT AVG(comfort_rating) FROM CombinedData);
```

# Q.N6

```python
%%sql
WITH AvgStarRatingPerSize AS (
    SELECT size, AVG(star_rating) AS avg_star_rating
    FROM your_table
    GROUP BY size
)

SELECT t1.product_id, t1.size, t1.star_rating
FROM your_table t1
JOIN AvgStarRatingPerSize t2 ON t1.size = t2.size
WHERE t1.star_rating > t2.avg_star_rating;
```
